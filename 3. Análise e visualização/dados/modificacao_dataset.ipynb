```python
import pandas as pd
```

```python
df_estudantes = pd.read_csv("/content/student_habits_performance.csv")
df_estudantes.info()
df_estudantes.columns
```

```python
colunas_originais = ['student_id', 'age', 'gender', 'study_hours_per_day',
       'social_media_hours', 'netflix_hours', 'part_time_job',
       'attendance_percentage', 'sleep_hours', 'diet_quality',
       'exercise_frequency', 'parental_education_level', 'internet_quality',
       'mental_health_rating', 'extracurricular_participation', 'exam_score']

# Novas colunas em português
colunas_novas = ['id_estudante', 'idade', 'sexo', 'horas_estudo_dia',
       'horas_redes_sociais', 'horas_netflix', 'trabalho_meio_periodo',
       'percentual_presenca', 'horas_sono', 'qualidade_dieta',
       'freq_exercicios', 'nivel_educacao_pais', 'qualidade_internet',
       'avaliacao_saude_mental', 'participacao_extracurricular', 'nota_exame']

# Mapeando as colunas antigas para as novas
mapeamento_colunas = dict(zip(colunas_originais, colunas_novas))

# Renomeando as colunas
df_estudantes = df_estudantes.rename(columns=mapeamento_colunas)

print(df_estudantes.columns.tolist())
df_estudantes.head()
```

```python
traducoes = {
    'Male': 'masculino',
    'Female': 'feminino',

    'Yes': 'sim',
    'No': 'não',

    'Poor': 'ruim',
    'Fair': 'razoável',
    'Good': 'bom',
    'Average': 'média',

    'High School': 'ensino médio',
    'Bachelor': 'graduação',
    'Master': 'mestrado',

    'Other': 'outro'
}

# Traduzindo as categorias
df_estudantes.replace(traducoes, inplace=True)
df_estudantes.head()
```

```python
# Verificação
for coluna in df_estudantes.columns:
    print(f"Valores únicos em '{coluna}':")
    print(df_estudantes[coluna].unique())
    print("\n")
```

```python
# Preenchendo valor faltante
df_estudantes['nivel_educacao_pais'] = df_estudantes['nivel_educacao_pais'].fillna("não informado")
```

```python
df_estudantes.to_csv("habitos-estudantis_performance-academica.csv", index=False)
```
